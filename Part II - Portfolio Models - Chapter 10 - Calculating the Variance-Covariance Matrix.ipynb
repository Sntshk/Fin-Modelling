```python
excess_returns = monthly_returns.iloc[1: , 1:] - monthly_returns_means
excess_returns.tail()
```

```python
var_cov_matrix = np.dot(excess_returns.iloc[:,:-1].T,
                        excess_returns.iloc[:,:-1]) / (excess_returns['MCD'].count() - 1)
```

```python
#using numpy cov
df = monthly_returns.iloc[1:,1:-1]
n = len(df.columns)
```

```python
matrix = pd.DataFrame(columns=df.columns[0:], index=df.columns[0:])
for i in range(0, n):
    for j in range(0, n):
        if i == j:
            matrix.iloc[i, j] = np.cov(df.iloc[:,i],df.iloc[:,j])[0][0]
        else:
            matrix.iloc[i, j] = np.cov(df.iloc[:,i],df.iloc[:,j])[0][1]
```

#### Computing the Global Minimum Variance Portfolio (GMVP)

```python
unitrowvector = np.ones(len(matrix.columns))
unitrowvector
```

```python
GMVP_as_row = np.dot(unitrowvector, matrix) / sum(np.dot(unitrowvector, matrix))
GMVP_as_row
```

```python
GMVP_as_column = np.array([GMVP_as_row]).T
GMVP_as_column
```

```python
#GMVP statistics
#Mean
sum(x*y for x,y in zip(GMVP_as_row, monthly_returns_means[0:-1]))
```

```python
#Variance
np.dot(np.dot(GMVP_as_row, matrix), GMVP_as_column)[0]
```

```python
#Sigma
np.sqrt(np.dot(np.dot(GMVP_as_row, matrix), GMVP_as_column)[0])
```

#### Four Alternatives to the Sample Variance-Covariance: The Single-Index Model (SIM)

```python
from scipy.stats import linregress
```

```python
assetdata = monthly_returns.iloc[1:,1:-1]
marketdata = monthly_returns.iloc[1:,-1]
sim = pd.DataFrame(columns=df.columns[0:], index=df.columns[0:])
for i in range(0, n):
    for j in range(0, n):
        if i == j:
            sim.iloc[i, j] = np.var(assetdata.iloc[:,i], ddof=1)
        else:
            sim.iloc[i, j] = linregress(marketdata, assetdata.iloc[:,i]
                                       ).slope * linregress(marketdata, assetdata.iloc[:,j],
                                                           ).slope * np.var(marketdata, ddof=1)
```

```python
sim
```

#### Alternatives to the Sample Variance-Covariance: Constant Correlation

```python
corr = 0.2
```

```python
const = pd.DataFrame(columns=df.columns[0:], index=df.columns[0:])
for i in range(0, n):
    for j in range(0, n):
        if i == j:
            const.iloc[i, j] = np.var(assetdata.iloc[:,i], ddof=1)
        else:
            const.iloc[i, j] = corr * np.std(assetdata.iloc[:,i], ddof=1) * np.std(assetdata.iloc[:,j], ddof=1)
```

```python
def CallVolatility(Stock, Exercise, Time, Interest, Target):
    High = 2
    Low = 0
    while (High - Low) > 0.0001:
        if BSCall(Stock, Exercise, Time, Interest, (High + Low) / 2) > Target:
            High = (High + Low) / 2
        else:
            Low = (High + Low) / 2
    CallVolatility = (High + Low) / 2
    return CallVolatility
```

```python
from scipy.stats import norm
```

```python
def BSCall(Stock, Exercise, Time, Interest, sigma):
    BSCall = Stock * norm.cdf(dOne(Stock, Exercise, Time, Interest, sigma)
                    ) - Exercise * np.exp(-Time * Interest) * norm.cdf(dTwo(Stock, Exercise, Time, Interest, sigma))
    return BSCall
```

```python
def dOne(Stock, Exercise, Time, Interest, sigma):
    dOne = (np.log(Stock / Exercise) + Interest * Time) / (sigma * np.sqrt(Time)) + 0.5 * sigma * np.sqrt(Time)
    return dOne
```

```python
def dTwo(Stock, Exercise, Time, Interest, sigma):
    dTwo = dOne(Stock, Exercise, Time, Interest, sigma) - sigma * np.sqrt(Time)
    return dTwo
```
